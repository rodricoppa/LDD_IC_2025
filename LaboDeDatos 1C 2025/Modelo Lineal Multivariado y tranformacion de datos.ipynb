```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Ejercicio 1
macroFull = pd.read_csv(r"C:\Users\Usuario\Downloads\macro_full_columns.csv")
macroFull
```

```python
macroFull.columns
```

```python
oferta_cols = [col for col in macroFull.columns if col.endswith(".oferta")]
macroFull_oferta = macroFull[oferta_cols]
macroFull_oferta
```

```python
# Veo si hay datos faltantes
faltantes = macroFull_oferta.isnull().sum()
faltantes
```

```python
print("\nDatos faltantes por columna:\n", faltantes)
print("\n¿Hay datos faltantes?:", macroFull_oferta.isnull().values.any())
```

```python
# b)
y = macroFull_oferta["PBI_a_precios_de_mercado.oferta"]
X = macroFull_oferta.drop(columns=["PBI_a_precios_de_mercado.oferta"])
y
```

```python
X
```

```python
print("Shape de X:", X.shape)
print("Shape de y:", y.shape)
print("\nPrimeras filas de X:")
print(X.head())
print("\nPrimeras filas de y:")
print(y.head())
```

```python
# c)
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X, y)
```

```python
y_pred = modelo.predict(X)
y_pred
```

```python
# Primeras predicciones
print("Primeras predicciones:")
print(y_pred[:5])
```

```python
# d)
import matplotlib.pyplot as plt
```

```python
# Grafico
plt.figure(figsize=(10,6))
plt.plot(y.index, y, label='PBI Real', marker='o')
plt.plot(y.index, y_pred, label='PBI Predicho', linestyle='--', marker='x')
plt.xlabel('Año')
plt.ylabel('PBI')
plt.title('Comparación entre PBI real y predicho')
plt.legend()
plt.grid(True)
plt.show()
```
