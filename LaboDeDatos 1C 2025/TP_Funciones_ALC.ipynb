```python
#FUNCIONES INDEPENDIENTES

def haycero(A):
    largo = np.shape(A)[0]
    k=False
    for i in range(largo):
        if A[i][i]==0:
            k=True
    return k


def posicionCero(A):
    i=0
    
    while (A[i][i]!=0):
        i=i+1
    return i

def distintoCero(A,i):
    largo = np.shape(A)[0]
    k=0

    while(A[k][i]==0 or A[k][:i]!=np.zeros(i)) :
        k = k+1
    return k

def primer_distintoCero(A,i):

    k=0

    while(A[k][i]==0):
        k = k+1
    return k

def intercambiarFilas(A,i,k):
        B=np.copy(A)
        A[i]=B[k]
        A[k]=B[i]
        return A

def ampliarMatriz(A):
    largo= np.shape(A)[0]
    B= np.eye(largo)
    C=np.hstack((A,B))
    return C

def traspuesta(A):
    largo=np.shape(A)[0]
    F=np.zeros((largo,largo))
    for i in range(largo):
        for j in range(largo):
            F[j][i] = A[i][j]  # Intercambiar filas por columnas
    return F
```

```python
#PIVOTEOS

def pivoteo(A):

    while(haycero(A)==True):
        intercambiarFilas(A,posicionCero(A),distintoCero(A,posicionCero(A)))
    
    return A

def primer_pivoteo(A):

    largo = np.shape(A)[0]
    Id=np.eye(largo)

    while(haycero(A)==True):
        Id = intercambiarFilas(Id,posicionCero(A),primer_distintoCero(A,posicionCero(A)))
        A = intercambiarFilas(A,posicionCero(A),primer_distintoCero(A,posicionCero(A)))
    
    return A,Id
```

```python
#CALCULAR L, U Y SIN CEROS

def sinCeros(A): 

    largo = np.shape(A)[0]
    A_pivoteada = A
    PrimerPivote = np.eye(largo)

    while(haycero(A)==True):
        PrimerPivote=primer_pivoteo(A)[1]
        A_pivoteada=primer_pivoteo(A)[0]

    return A_pivoteada, PrimerPivote

def calcularL (A):
    
    Mi=np.eye(np.shape(A)[0],k=0)
    largo = np.shape(A)[0]

    for i in range (largo-1):
        Mj=np.eye(largo)
        M=np.eye(largo)
        A=pivoteo(A)
        
        for j in range (i+1,largo):
            
            M[j][i]= -A[j][i]/A[i][i]
            Mj[j][i]= A[j][i]/A[i][i]
        
        A=M@A
        Mi = Mi@Mj
    return Mi

def calcularU (A):
    
    for i in range (np.shape (A)[0]-1):
        M=np.eye(np.shape(A)[0],k=0)
        for j in range (i+1,np.shape(A)[0] ):
            pivoteo(A)
            M[j][i]= -A[j][i]/A[i][i]
        
        A=M@A
    return A
```

```python
#INVERSAS L Y U

def inversaL(L):
    largo =np.shape(L)[0]
    L=ampliarMatriz(L)
    for k in range(largo):
        # Normalizar el elemento diagonal
        L[k] = L[k] / L[k][k]

    for k in range(largo):
        for j in range (k+1, largo):
            for i in range (largo):
                A = L[j][k] #indice de multiplicacion
                B = L[k] #permutador
                L[j]=L[j]-(B*A)
    return L[:,largo:]

def inversaU(U):

    U=traspuesta(U)
    U=inversaL(U)
    U=traspuesta(U)

    return U

def inversaLU(L, U):
    Inv = []
    Inv=inversaU(U)@inversaL(L)
    return Inv
```

```python
#CALCULAR LU

def calcularLU(A):

    B = sinCeros(A)[1]
    A = sinCeros(A)[0]

    L, U = [],[]
    L=calcularL(A)
    U=calcularU(A)
    P=B@A@(inversaLU(calcularL(A),calcularU(A))) 
    P=np.round(P)

    return L, U, P
```
