```python
resultado = (
    penguins[penguins["island"] != "Biscoe"]
    .loc[:, lambda df : df.columns[df.columns.get_loc("species") : df.columns.get_loc("body_mass_g") + 1]]
    .rename(columns ={"species" : "especie_pinguino"})
    .groupby("especie_pinguino")
    .mean(numeric_only = True)
    .filter(like = "length")
    .add_suffix("_mean")
)
resultado
```

```python
# Ejercicio 15
penguins_filtrado = penguins[penguins['island'] != 'Biscoe']
```

```python
penguins_filtrado = penguins_filtrado.loc[:, "species":"body_mass_g"]
penguins_filtrado
```

```python
penguins_filtrado = penguins_filtrado.rename(columns = {"species":"especie_pinguino"})
penguins_filtrado
```

```python
penguins_agrupados = penguins_filtrado.groupby("especie_pinguino")
penguins_agrupados
```

```python
columnas_length = [col for col in penguins_filtrado.columns if "length" in col]
columnas_length
```

```python
resultado = penguins_agrupados[columnas_length].agg("mean")
resultado
```

```python
resultado = resultado.rename(columns = {col : col + "_mean" for col in columnas_length})
resultado
```
